```python
import pandas as pd
import numpy as np
import synapseclient as sc
```

```python
syn = sc.login()
```

```python
entity = syn.get("syn20817213")
sfm_data = pd.read_csv(entity["path"], index_col = 0)
```

```python
def fill(params):
    try:
        return float(params)
    except:
        return np.NaN
    
for feat in [feat for feat in sfm_data.columns if "sfm" in feat]:
    sfm_data[feat] = sfm_data[feat].apply(fill)
```

```python
sfm_mean_data = sfm_data.groupby("healthCode").mean()[[feat for feat in sfm_data.columns if "sfm" in feat]].reset_index()
sfm_min_data = sfm_data.groupby("healthCode").min()[[feat for feat in sfm_data.columns if "sfm" in feat]].reset_index()
```

```python
filtered_entity = syn.get("syn8381056")
filtered_healthcode_data = pd.read_csv(filtered_entity["path"], sep = "\t")
filtered_healthcode_list = list(filtered_healthcode_data["healthCode"])
```

```python
walking_data = pd.read_csv(r"C:\Users\atedi\OneDrive\Documents\SageBionetworks\local_data\walking_data.csv", index_col=0)
balance_data = pd.read_csv(r"C:\Users\atedi\OneDrive\Documents\SageBionetworks\local_data\balance_data.csv", index_col=0)
```

```python
mean_walking_data = walking_data.groupby("healthCode").mean()[["no_of_steps userAccel_walking_features_AA", "speed_of_gait userAccel_walking_features_AA"]].reset_index()
max_walking_data = walking_data.groupby("healthCode").max()[["count_freeze_index userAccel_walking_features_AA"]].reset_index()
```

```python
data = pd.merge(mean_walking_data, max_walking_data, how = "left", on = "healthCode")
data = pd.merge(data, sfm_mean_data, how = "left", on = "healthCode")
data = pd.merge(data, sfm_min_data, how = "left", on = "healthCode")
data = pd.merge(data, filtered_healthcode_data, how = "left", on = "healthCode")
```
